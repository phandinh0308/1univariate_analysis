```python
# %%
# =========================================================
# IMPORT THƯ VIỆN
# =========================================================

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# %%
# =========================================================
# LOAD DATA
# Data đã xử lý + feature engineering sẵn
# =========================================================

df = pd.read_excel(
    r"C:\Users\T480s\Documents\COLLECT_DATA\data_multivariate.xlsx"
)

# %%
# =========================================================
# HÀM TẠO CROSSTAB
# =========================================================

def tao_bang_cheo(
    df,
    row_var,
    col_var
):

    count_table = pd.crosstab(
        df[row_var],
        df[col_var]
    )

    pct_total_table = (
        pd.crosstab(
            df[row_var],
            df[col_var],
            normalize='all'
        ) * 100
    )

    pct_col_table = (
        pd.crosstab(
            df[row_var],
            df[col_var],
            normalize='columns'
        ) * 100
    )

    pct_row_table = (
        pd.crosstab(
            df[row_var],
            df[col_var],
            normalize='index'
        ) * 100
    )

    return {
        'count': count_table,
        'pct_total': pct_total_table,
        'pct_col': pct_col_table,
        'pct_row': pct_row_table
    }

# %%
# =========================================================
# FORMAT LABEL %
# =========================================================

def format_percent_label(table):

    labels = np.array([

        f'{value:.1f}%'

        for value in table.values.flatten()

    ]).reshape(table.shape)

    return labels

# %%
# =========================================================
# HÀM VẼ 4 HEATMAP
# =========================================================

def ve_heatmap_4_o(
    tables,
    row_var,
    col_var,
    title,
    figsize=(18,12),
    cmap='YlOrBr'
):

    fig, axes = plt.subplots(
        2,
        2,
        figsize=figsize
    )

    # =====================================================
    # COUNT
    # =====================================================

    sns.heatmap(
        tables['count'],
        annot=True,
        fmt='d',
        cmap=cmap,
        ax=axes[0,0]
    )

    axes[0,0].set_title(
        f'Count - {title}'
    )

    # =====================================================
    # % TỔNG THỂ
    # =====================================================

    sns.heatmap(
        tables['pct_total'],
        annot=format_percent_label(
            tables['pct_total']
        ),
        fmt='',
        cmap=cmap,
        ax=axes[0,1]
    )

    axes[0,1].set_title(
        'Percentage on Total Dataset'
    )

    # =====================================================
    # % THEO CỘT
    # =====================================================

    sns.heatmap(
        tables['pct_col'],
        annot=format_percent_label(
            tables['pct_col']
        ),
        fmt='',
        cmap=cmap,
        ax=axes[1,0]
    )

    axes[1,0].set_title(
        f'Percentage by {col_var}'
    )

    # =====================================================
    # % THEO HÀNG
    # =====================================================

    sns.heatmap(
        tables['pct_row'],
        annot=format_percent_label(
            tables['pct_row']
        ),
        fmt='',
        cmap=cmap,
        ax=axes[1,1]
    )

    axes[1,1].set_title(
        f'Percentage by {row_var}'
    )

    # =====================================================
    # FORMAT CHUNG
    # =====================================================

    for ax in axes.flat:

        ax.set_xlabel(col_var)

        ax.set_ylabel(row_var)

        ax.tick_params(
            axis='x',
            rotation=90
        )

    plt.tight_layout()

    plt.show()

# %%
# =========================================================
# HÀM GHÉP:
# TẠO BẢNG + VẼ HEATMAP
# =========================================================

def xu_ly_heatmap(
    df,
    row_var,
    col_var,
    title
):

    tables = tao_bang_cheo(
        df=df,
        row_var=row_var,
        col_var=col_var
    )

    ve_heatmap_4_o(
        tables=tables,
        row_var=row_var,
        col_var=col_var,
        title=title
    )

# %%
# =========================================================
# DANH SÁCH CÁC CẶP BIẾN
# =========================================================

heatmap_configs = [

    {
        'row_var': 'loan_amount_group',
        'col_var': 'term',
        'title': 'Loan Amount vs Term'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'interestrate_group',
        'title': 'Loan Amount vs Interest Rate'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'dti_total_group',
        'title': 'Loan Amount vs DTI Total'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'residual_income_group',
        'title': 'Loan Amount vs Residual Income'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'emi_total_group',
        'title': 'Loan Amount vs EMI Total'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'age_bin',
        'title': 'Loan Amount vs Age'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'gender_label',
        'title': 'Loan Amount vs Gender'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'customer_type',
        'title': 'Loan Amount vs Customer Type'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'province_region',
        'title': 'Loan Amount vs Region'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'vc_score_bin',
        'title': 'Loan Amount vs VC Score'
    },

    {
        'row_var': 'loan_amount_group',
        'col_var': 'utilization_group',
        'title': 'Loan Amount vs Utilization'
    }

]

# %%
# =========================================================
# LOOP AUTO VẼ HEATMAP
# =========================================================

for config in heatmap_configs:

    print(
        f'\n========== '
        f'{config["title"].upper()} '
        f'=========='
    )

    xu_ly_heatmap(

        df=df,

        row_var=config['row_var'],

        col_var=config['col_var'],

        title=config['title']

    )
```
