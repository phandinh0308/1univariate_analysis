```python
# %%
# =========================================================
# IMPORT THƯ VIỆN
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# %%
# =========================================================
# LOAD DATA
# =========================================================

df = pd.read_excel(
    r'C:\Users\T480s\Documents\COLLECT_DATA\data_processed.xlsx'
)

# %%
# =========================================================
# HÀM CHIA BIN
# =========================================================

def chia_bin(
    df,
    col,
    bins,
    labels,
    new_col,
    right=True,
    include_lowest=True
):

    # -----------------------------------------------------
    # TỰ ĐỘNG ĐÁNH SỐ THỨ TỰ LABEL
    # -----------------------------------------------------

    labels = [

        f'{i+1}. {label}'

        for i, label in enumerate(labels)

    ]

    # -----------------------------------------------------
    # TẠO BIẾN SAU KHI CHIA BIN
    # -----------------------------------------------------

    df[new_col] = pd.cut(

        df[col],

        bins=bins,

        labels=labels,

        right=right,

        include_lowest=include_lowest

    )

    # -----------------------------------------------------
    # CHUYỂN CATEGORY CÓ THỨ TỰ
    # -----------------------------------------------------

    df[new_col] = pd.Categorical(

        df[new_col],

        categories=labels,

        ordered=True

    )

# %%
# =========================================================
# HÀM TẠO SUMMARY
# =========================================================

def tao_summary(
    df,
    col
):

    summary = pd.DataFrame({

        'count': (

            df[col]

            .value_counts(
                dropna=False
            )

            .sort_index()

        ),

        'percent (%)': (

            df[col]

            .value_counts(
                normalize=True,
                dropna=False
            )

            .sort_index()

            * 100

        )

    })

    return summary.round(2)

# %%
# =========================================================
# HÀM VẼ BIỂU ĐỒ
# =========================================================

def ve_bieu_do(
    summary,
    title,
    rotation=90,
    figsize=(16,5),
    color_count='steelblue',
    color_percent='orange'
):

    fig, ax = plt.subplots(

        1,
        2,

        figsize=figsize

    )

    # =====================================================
    # BIỂU ĐỒ COUNT
    # =====================================================

    bars1 = ax[0].bar(

        summary.index.astype(str),

        summary['count'],

        color=color_count

    )

    ax[0].set_title(
        f'{title} - Count'
    )

    ax[0].tick_params(

        axis='x',
        rotation=rotation

    )

    # ----- Hiển thị label count -----

    for bar in bars1:

        ax[0].text(

            bar.get_x() + bar.get_width()/2,

            bar.get_height(),

            int(bar.get_height()),

            ha='center',
            va='bottom'

        )

    # =====================================================
    # BIỂU ĐỒ PERCENTAGE
    # =====================================================

    bars2 = ax[1].bar(

        summary.index.astype(str),

        summary['percent (%)'],

        color=color_percent

    )

    ax[1].set_title(
        f'{title} - Percentage'
    )

    ax[1].tick_params(

        axis='x',
        rotation=rotation

    )

    # ----- Hiển thị label percentage -----

    for bar in bars2:

        ax[1].text(

            bar.get_x() + bar.get_width()/2,

            bar.get_height(),

            f'{bar.get_height():.1f}%',

            ha='center',
            va='bottom'

        )

    plt.tight_layout()

    plt.show()

# %%
# =========================================================
# HÀM GHÉP:
# CHIA BIN + SUMMARY + VẼ CHART
# =========================================================

def xu_ly_va_ve(
    df,
    col,
    title,
    bins,
    labels,
    new_col
):

    # -----------------------------------------------------
    # CHIA BIN
    # -----------------------------------------------------

    chia_bin(

        df=df,

        col=col,

        bins=bins,

        labels=labels,

        new_col=new_col

    )

    # -----------------------------------------------------
    # TẠO SUMMARY
    # -----------------------------------------------------

    summary = tao_summary(

        df=df,

        col=new_col

    )

    # -----------------------------------------------------
    # HIỂN THỊ BẢNG
    # -----------------------------------------------------

    display(summary)

    # -----------------------------------------------------
    # VẼ BIỂU ĐỒ
    # -----------------------------------------------------

    ve_bieu_do(

        summary=summary,

        title=title

    )

# %%
# =========================================================
# DANH SÁCH CẤU HÌNH BIẾN
# =========================================================

configs = [

    # =====================================================
    # LOAN AMOUNT
    # =====================================================

    {
        'col': 'loan_amount',

        'title': 'Loan Amount',

        'bins': [

            -float('inf'),

            10_000_000,
            15_000_000,
            20_000_000,

            30_000_000,
            40_000_000,
            50_000_000,

            float('inf')

        ],

        'labels': [

            '9-10M',
            '11-15M',
            '16-20M',

            '21-30M',
            '31-40M',
            '41-50M',

            '>50M'

        ],

        'new_col': 'loan_amount_bin'

    },

    # =====================================================
    # AGE
    # =====================================================

    {
        'col': 'age',

        'title': 'Age',

        'bins': [

            18,
            22,
            26,
            30,
            40,

            float('inf')

        ],

        'labels': [

            '18-22',
            '22-26',
            '26-30',
            '30-40',
            '>40'

        ],

        'new_col': 'age_bin'

    },

    # =====================================================
    # EMI TOTAL
    # =====================================================

    {
        'col': 'emi_total',

        'title': 'EMI Total',

        'bins': [

            -float('inf'),

            1_500_000,
            2_200_000,
            3_000_000,

            5_000_000,
            10_000_000,
            50_000_000,

            float('inf')

        ],

        'labels': [

            '≤1.5M',
            '1.5-2.2M',
            '2.2-3M',

            '3-5M',
            '5-10M',
            '10-50M',

            '>50M'

        ],

        'new_col': 'emi_total_bin'

    },

    # =====================================================
    # DTI TOTAL
    # =====================================================

    {
        'col': 'dti_total',

        'title': 'DTI Total',

        'bins': [

            -float('inf'),

            15,
            20,
            25,

            35,
            50,
            100,

            float('inf')

        ],

        'labels': [

            '≤15%',
            '15-20%',
            '20-25%',

            '25-35%',
            '35-50%',
            '50-100%',

            '>100%'

        ],

        'new_col': 'dti_total_bin'

    },

    # =====================================================
    # SCORE
    # =====================================================

    {
        'col': 'score',

        'title': 'Score',

        'bins': [

            -float('inf'),

            500,
            550,
            600,

            650,
            700,
            750,

            float('inf')

        ],

        'labels': [

            '<500',
            '500-550',
            '550-600',

            '600-650',
            '650-700',
            '700-750',

            '>750'

        ],

        'new_col': 'score_bin'

    }

]

# %%
# =========================================================
# LOOP TỰ ĐỘNG CHẠY TOÀN BỘ BIẾN
# =========================================================

for config in configs:

    print(
        f"\n========== {config['title'].upper()} =========="
    )

    xu_ly_va_ve(

        df=df,

        col=config['col'],

        title=config['title'],

        bins=config['bins'],

        labels=config['labels'],

        new_col=config['new_col']

    )

# %%
# =========================================================
# KIỂM TRA CÁC BIẾN SAU KHI CHIA BIN
# =========================================================

bin_cols = [

    config['new_col']

    for config in configs

]

display(
    df[bin_cols].head()
)
```
