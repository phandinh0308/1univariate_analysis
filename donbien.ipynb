```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel(r'C:\Users\T480s\Documents\COLLECT_DATA\data_processed.xlsx')
```

```python
#Biến loan_amount
#Chia các bin
bins = [
    -float('inf'),
    10_000_000,
    15_000_000,
    20_000_000,
    30_000_000,
    40_000_000,
    50_000_000,
    float('inf')
]

labels = [
    '1. 9-10M',
    '2. 11-15M',
    '3. 16-20M',
    '4. 21-30M',
    '5. 31-40M',
    '6. 41-50M',
    '7. >50M'
]

df['loan_bin'] = pd.cut(
    df['loan_amount'],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)
#Đếm số lượng mỗi nhóm
count_data = df['loan_bin'].value_counts().sort_index()
percent_data = count_data / count_data.sum() * 100
fig, ax = plt.subplots(1, 2, figsize=(16, 5))
#Vẽ biểu đồ
for i, (data, title, ylabel, fmt) in enumerate([
    (count_data, 'Loan Amount - Count', 'Count', '{:.0f}'),
    (percent_data, 'Loan Amount - Percentage', '%', '{:.1f}%')
]):
    ax[i].bar(data.index.astype(str), data.values)
    ax[i].set_title(title)
    ax[i].set_xlabel('Loan Amount Group')
    ax[i].set_ylabel(ylabel)
    ax[i].tick_params(axis='x', rotation=45)

    # Data label
    for x, y in enumerate(data.values):
        ax[i].text(x, y, fmt.format(y), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
#Biến score và vc_score
#Hàm vẽ biểu đồ
def plot_ordinal(summary, col_name):
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # COUNT
    bars1 = axes[0].bar(summary.index.astype(str), summary['count'])
    axes[0].set_title(f"{col_name} - Count")
    axes[0].tick_params(axis='x', rotation=45)

    for b in bars1:
        axes[0].text(b.get_x() + b.get_width()/2,
                     b.get_height(),
                     int(b.get_height()),
                     ha='center', va='bottom')

    # PERCENT
    bars2 = axes[1].bar(summary.index.astype(str), summary['percent (%)'])
    axes[1].set_title(f"{col_name} - Percentage (%)")
    axes[1].tick_params(axis='x', rotation=45)

    for b in bars2:
        axes[1].text(b.get_x() + b.get_width()/2,
                     b.get_height(),
                     f"{b.get_height():.2f}%",
                     ha='center', va='bottom')

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.25)
    plt.show()


# Danh sách các biến
ordinal_cols = ['score_bin', 'vc_score_bin']

# Loop qua từng biến
for col in ordinal_cols:
    
    # Tính Count & % ---
    summary = df[col].value_counts().sort_index().to_frame(name='count')
    summary['percent (%)'] = summary['count'] / summary['count'].sum() * 100

    # Vẽ biểu đồ
    plot_ordinal(summary, col)
```

```python
#Biến term và month
def plot_distribution(df, col):
    summary = df[col].value_counts().sort_index().to_frame(name='count')
    summary['percent (%)'] = summary['count'] / summary['count'].sum() * 100
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))

    # COUNT
    bars1 = ax[0].bar(summary.index.astype(str), summary['count'])
    ax[0].set_title(f'{col} - Count')
    ax[0].set_xlabel(col)
    ax[0].set_ylabel('Count')

    for b in bars1:
        ax[0].text(b.get_x() + b.get_width()/2,
                   b.get_height(),
                   int(b.get_height()),
                   ha='center', va='bottom')

    # PERCENT
    bars2 = ax[1].bar(summary.index.astype(str), summary['percent (%)'])
    ax[1].set_title(f'{col} - Percentage')
    ax[1].set_xlabel(col)
    ax[1].set_ylabel('%')

    for b in bars2:
        ax[1].text(b.get_x() + b.get_width()/2,
                   b.get_height(),
                   f"{b.get_height():.1f}%",
                   ha='center', va='bottom')

    plt.tight_layout()
    plt.show()


# Chạy cho 2 biến
for col in ['term', 'month']:
    plot_distribution(df, col)
```
